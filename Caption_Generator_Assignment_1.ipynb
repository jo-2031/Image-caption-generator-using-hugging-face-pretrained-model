```python
from PIL import Image
from transformers import VisionEncoderDecoderModel, ViTFeatureExtractor,PreTrainedTokenizerFast
import requests 
```

```python
model1 = VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
feature_extractor1 = ViTFeatureExtractor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
tokenizer1 = PreTrainedTokenizerFast.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
```

```python
#inference function
def vitgpt2(img):
  pixel_values=feature_extractor1(images=img,return_tensors="pt").pixel_values
  encoder_ouputs = model1.generate(pixel_values.to('cpu'),num_beams = 5)
  generator_sentences = tokenizer1.batch_decode(encoder_ouputs,skip_special_tokens=True)

  return(generator_sentences[0].split('.')[0])
```

```python
import gradio as gr
```

```python
inputs = [
    gr.inputs.Image(type= 'pil',label='original Image')
]

outputs = [
    gr.outputs.Textbox(label='Caption')
]


title  = 'Image Caption Generator using VIT GPT2 '
description = "Build a caption generator using Hugging Face VIt GPT-2  is a natural language processing model that can automatically generate captions for images or videos using the Hugging Face implementation of the Vision Transformer (VIt) and GPT-2, which are two powerful deep learning models.VIt Gp2 is pretrained model for Build a aption generator"
articles = "<a href = 'https://huggingface.co/nlpconnect/vit-gpt2-image-captioning'> Model where used for the caption generator"
examples = [
    ['front-left-side-47.jpg']
]

gr.Interface(
    vitgpt2,
    inputs,
    outputs,
    title = title,
    description = description,
    example = examples,
    theme = "huggingface",

).launch(debug = True,enable_queue=True)
```
